```python
from folium.plugins import HeatMap

map_x = df['위도']
map_y = df['경도']
map_t = df['업체명']


coords = []
for i in range(len(df)):
      x = map_x[i]
      y = map_y[i]
      coords.append([x, y])

title=[]    
for i in range(len(df)):
    t = map_t[i]
    title.append([t])



def allroute():
    m = folium.Map(location=(37.5776087830657, 126.976896737645), zoom_start=10)
    
    heatMap = HeatMap(zip(map_x,map_y),
                   min_opacity=0.1,
                   max_val=0.5,
                   radius=20, blur=5,
                   max_zoom=5,color='red')

    m.add_child(heatMap)
    
    return m
```

정규표현식

```python
def extract_word(text):
    hangul = re.compile('[^가-힣]') 
    result = hangul.sub(' ', text) 
    return result
```

```python
df_제조업['업종명'] = df_제조업['업종명'].apply(lambda x:extract_word(x))
df_정보처리['업종명'] = df_정보처리['업종명'].apply(lambda x:extract_word(x))
df_기타['업종명'] = df_기타['업종명'].apply(lambda x:extract_word(x))
df_연구개발['업종명'] = df_연구개발['업종명'].apply(lambda x:extract_word(x))
df_도소매업['업종명'] = df_도소매업['업종명'].apply(lambda x:extract_word(x))
df_건설['업종명'] = df_건설['업종명'].apply(lambda x:extract_word(x))
df_농어임['업종명'] = df_농어임['업종명'].apply(lambda x:extract_word(x))

# 모든 데이터프레임에 정규표현식을 적용하여 한글만 남기기
```

##1.제조업 워드클라우드

```python
word_제조업 = []
stop_words = ['제조업', '기타']
twt = Twitter()

# 형태소 추출
for n in range(len(df_제조업)):
  text = df_제조업['업종명'].iloc[n]
  words = twt.pos(text)
  
  # 불용어를 제외한 명사만 추출하기
  for i, j in words:
      if j == 'Noun':
        if i not in stop_words: 
          word_제조업.append(i)

# 한 글자 제거 
word_제조업 = [x for x in word_제조업 if len(x)>1]

# count 하기
counts_제조업 = Counter(word_제조업)
```
